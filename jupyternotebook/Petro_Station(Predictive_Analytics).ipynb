# Petro-Station (Predictive Analytics)

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns

print("Setup Complete")
```

# Load to Dataframe

```python
j = pd.read_excel('journal (1).xlsx')
j
```

# Data Exploration

```python
jd = pd.read_excel('journal_detail.xlsx')
jd
```

```python
df = pd.concat([j,jd])
df
df.head()
```

# Choosing Features

```python
jd_features = ['goods_amout','tax_amout']
y = jd.total
X = jd[jd_features]
X.describe()
from sklearn.tree import DecisionTreeRegressor
jd_model = DecisionTreeRegressor(random_state=1)
jd_model.fit(X, y)
```

# Some Predictions

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(jd_model.predict(X.head()))
```

# Model Validation

```python
from sklearn.metrics import mean_absolute_error
predicted = jd_model.predict(X)
mean_absolute_error(y, predicted)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
jd_model = DecisionTreeRegressor()
jd_model.fit(train_X, train_y)
val_predictions = jd_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

# Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```
