```python
import pandas as pd

pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns

print("Setup Complete")
```

```python
Lemonade_filepath = "/content/Lemonade.csv"
Lemonade_data = pd.read_csv(Lemonade_filepath, index_col="Date", parse_dates=True)
Lemonade_data
```

```python
list(Lemonade_data.columns)
plt.figure(figsize=(14,6))
sns.lineplot(data=Lemonade_data['Sales'], label="Sales")
sns.lineplot(data=Lemonade_data['Temperature'], label="Temperature")
plt.xlabel("Date")
```

```python
plt.figure(figsize=(8,5))
Lemonade_data['Average Revenue'] = Lemonade_data['Price'] * Lemonade_data['Sales']
plt.bar(Lemonade_data['Day'], Lemonade_data['Average Revenue'])
plt.title('Average Revenue')
plt.show()
```

```python
import matplotlib.pyplot as plt
Lemonade_data['Revenue'] = Lemonade_data['Sales'] * Lemonade_data['Price']
plt.pie(Lemonade_data['Revenue'], labels=Lemonade_data['Day'] )
plt.axis('equal')
plt.title('Average Revenue by Day')
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.title("Sales")
plt.scatter(Lemonade_data["Flyers"], Lemonade_data["Sales"])
plt.xlabel("Sales")
plt.ylabel("Flyers")
plt.show()
```
